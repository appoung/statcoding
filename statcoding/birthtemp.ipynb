```python
import csv

import matplotlib.pyplot as plt
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
result = []
for row in data:
    if row[4] != '':
        result.append(float(row[4]))
plt.figure(figsize=(50, 5))
plt.plot(result)
plt.title('Seoul Temperature')
plt.show()
f.close()
```
